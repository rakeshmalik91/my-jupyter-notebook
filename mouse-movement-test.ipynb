```python
import matplotlib.pyplot as plt
import numpy as np

def plot_diagram():
    fig, ax = plt.subplots(figsize=(6,6))
    
    # Draw the circular plate
    circle = plt.Circle((0, 0), 1, color='blue', fill=False, linewidth=2, linestyle='dashed')
    ax.add_patch(circle)
    
    # Define mouse base shape (trapezium)
    mouse_top_width, mouse_bottom_width, mouse_height = 0.4, 0.6, 0.3  # Approximate dimensions
    
    # Function to create a trapezium
    def trapezium(x, y, top_width, bottom_width, height):
        trap_x = [x - bottom_width/2, x + bottom_width/2, x + top_width/2, x - top_width/2, x - bottom_width/2]
        trap_y = [y - height/2, y - height/2, y + height/2, y + height/2, y - height/2]
        return trap_x, trap_y
    
    # Mouse positions inside the circular path
    radius_offset = 0.8  # Mouse moves slightly inside the circle
    angles = np.linspace(0, 2*np.pi, 100, endpoint=False)
    for angle in angles:
        x_offset, y_offset = radius_offset * np.cos(angle), radius_offset * np.sin(angle)
        trap_x, trap_y = trapezium(x_offset, y_offset, mouse_top_width, mouse_bottom_width, mouse_height)
        ax.plot(trap_x, trap_y, color='red')
        
        # Mark tracking sensor (assumed slightly off-center)
        sensor_x = x_offset + 0.15 * np.cos(angle + np.pi/2)
        sensor_y = y_offset + 0.15 * np.sin(angle + np.pi/2)
        ax.plot(sensor_x, sensor_y, 'ko', markersize=5, label='Tracking Sensor' if angle == 0 else "")
    
    # Labels & arrows
    ax.arrow(0.9, 0, -0.2, 0.2, head_width=0.05, head_length=0.05, fc='black', ec='black')
    ax.text(1.0, 0, "Motion Direction", fontsize=10)
    
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.legend()
    plt.title("Motion of a Trapezium-Shaped Mouse Inside a Circular Plate")
    plt.show()

plot_diagram()
```
