```python
# Input

X = {0,1,2,3,4,5,6}
c = 3
lmda = 1

m = len(X)
r = int(lmda * (m - 1) / (c - 1))
n = int(m * r / c)
```

```python
import time
import random
from math import comb
from itertools import combinations

print("m =", m, ", X =", X)
print("c =", c, ", λ =", lmda)
print("r =", r, ", n =", n)
print('\nGoing to search from', comb(comb(m,c), n), 'results')
startTime = time.time()

iX = {value: index for index, value in enumerate(X)}
def isValidBIBD(blocks, lmda):
    counts = [[0 for x in range(m)] for x in range(m)]
    for block in blocks:
        for i in combinations(block, 2):
            i1 = iX[i[0]]
            i2 = iX[i[1]]
            if counts[i1][i2] >= lmda:
                return False
            counts[i1][i2] += 1
    for row in counts:
        for cell in row:
            if cell not in {0, lmda}:
                return False
    return True

B=[]
allCombinations = list(combinations(X, c))
# random.shuffle(allCombinations)                                # just to randomly get a result
for possibleB in combinations(allCombinations, n):
    if isValidBIBD(possibleB, lmda):
        B = list(possibleB)
        print('B =', B)
        break

# for a0 in allCombinations:
#     a1toN = set(allCombinations)
#     a1toN.remove(a0)
#     print('\nfor A[0] = ', a0)
#     for possibleB in combinations(a1toN, n-1):
#         possibleB = (a0,) + possibleB
#         if isValidBIBD(possibleB, lmda):
#             B = list(possibleB)
#             print('B =', B)
            
endTime = time.time()
print("\nTime taken:", endTime - startTime, "seconds")
```

```python
import time
import sys

no_rows = m
no_cols = n
row_wt = r
col_wt = c

startTime = time.time()

all_combinations = list(combinations(range(no_rows), col_wt))

# This check lambda =1 for BIBD or not
# This function check each pair of elements are l times in the blocks or not
def check_pair_occurrences(subsets, l):
    all_pairs = set()
    for subset in subsets:
        all_pairs.update(combinations(subset, 2))
    pair_count = {pair: 0 for pair in all_pairs}
    for subset in subsets:
        for pair in combinations(subset, 2):
            pair_count[pair] += 1
#     print('pairs', pair_count)
    return all(count == l for count in pair_count.values())

# This check column weight is maintained or not.
def check_occurrences(sett, r):
    element_count = {}
    for s in sett:
        for e in s:
            if e in element_count:
                element_count[e] += 1
            else:
                element_count[e] = 1
    for e, count in element_count.items():
        if count != r:
            return False
    return True
    
# for i in range(len(all_combinations)):
#     B =  [all_combinations[i]]
#     print('i=',i)
#     print(B)
#     for j in range(len(all_combinations)):
#         if i != j:
#             subset = B + [all_combinations[j]]
#             print('j=',j)
#             print('subset = ',subset)
#             if check_pair_occurrences(subset, 1) == True:
#                 B = B + [all_combinations[j]]
#                 print('this=',B)
#                 print('---------------')
#                 if len(B) == no_cols:
#                     if check_occurrences(B, row_wt):
#                         print(B)
#                         print('----')
#                         sys.exit()

def run(B, remaining_combinations, result):
    if result != None:
        return result
    if len(B) == no_cols:
        if check_occurrences(B, row_wt):
            print('>> ', B)
            result = B
            return result
        else:
            print('   ', B)
            return None
    else:
        print('   ', B)
        for c in remaining_combinations:
            new_remaining_combinations = [x for x in remaining_combinations if x != c]
            new_B = B + [c]
            if check_pair_occurrences(new_B, lmda):
                result = run(new_B, new_remaining_combinations, result)
            if result != None:
                break
        return result

result = run([], list(all_combinations), None)

endTime = time.time()
print("\nTime taken:", endTime - startTime, "seconds")
```
